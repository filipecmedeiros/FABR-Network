```python
import numpy as np
import pandas as pd
from slugify import slugify
from datetime import datetime
from sqlalchemy import create_engine
engine = create_engine('postgresql://fabrnetwork:networkfootball2018@@@localhost:5432/FABR Network')
```

```python
positions = {
    0:"Nan",
    1:"Wide Receiver",
    2:"Quarterback",
    3:"Running Back",
    4:"Fullback",
    5:"Center",
    6:"Offensive Tackle",
    7:"Offensive Guard",
    8:"Tight End",
    9:"Kicker",
    10:"Punter",
    11:"Long Snapper",
    12:"Defensive Tackle",
    13:"Defensive End",
    14:"Linebacker",
    15:"Cornerback",
    16:"Strong Safety",
    17:"Free Safety",
    18:"Punt Returner",
    19:"Kick Returner",
    20:"Outside Linebacker",
    21:"Middle Linebacker",
}
chave = positions.keys()
valor = positions.values()
positions = dict(zip(valor,chave))
positions
```

```python
def rosterfy (players, team_id, start):
    players.name = players.name.apply(lambda x: x.title())
    players.number = players.number.fillna(value=0)
    players.number = players.number.apply(int)
    players.position_id = players.position_id.apply(str)
    players.position_id = players.position_id.apply(lambda x: x.title())
    players.position_id = players.position_id.apply(lambda x: positions[x])
    players.birthdate = players.birthdate.apply(lambda x: datetime.strptime(x, '%d/%m/%Y'))
    players.team_id = team_id
    players['created'] = '2018-10-26 22:34:20.463997-03'
    players['modified'] = '2018-10-26 22:34:20.463997-03'
    players['slug'] = ''
    players.slug = players.name.apply(lambda x: slugify(x))
    players['photo'] = 'players/photo/'+str(team_id)+'.jpg'
    
    players = players.drop_duplicates(subset='name', keep='first')
    players.index = np.arange(start, start+len(players))
    
    return players
```

```python
america = pd.read_csv('jogadorestimes/América Locomotiva.csv', header=None,
                      names=['name', 'number', 'birthdate', 'position_id', 'team_id'])
america = rosterfy(america, 1, 1)
america.to_sql('team_player', engine, index=True, index_label='id', if_exists='append')
america
```
